```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("person_data.csv", index_col=0)
```

```python
print(df)
```

```python
df.isnull().any()
```

```python
df = df.dropna(axis=0).reset_index(drop=True)
df
```

```python
df.info()
```

```python
data_height = df.height
data_weight = df.weight
```

```python
%matplotlib inline

import matplotlib.pyplot as plt

plt.scatter(data_height, data_weight)
plt.xlabel("height(cm)")
plt.ylabel("weight(kg)")
plt.show()
```

```python
X = np.array(data_height).reshape(len(data_height), 1)
y = np.array(data_weight).reshape(len(data_weight), 1)
```

```python
X
```

```python
y
```

```python
from sklearn import preprocessing

mm_scaled = preprocessing.MinMaxScaler()
X_train = mm_scaled.fit_transform(X)
y_train = mm_scaled.transform(y)

plt.scatter(X_train, y_train)
plt.xlabel('scaled-height')
plt.ylabel('scaled-weight')
plt.show()
```

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense

#  .
model = Sequential()

#    1   1 y=wx+b  .
model.add(Dense(1, input_shape=(1,)))

# Loss funtion Optimizer .
model.compile(loss='mse', optimizer='sgd', metrics=['accuracy']) 

# epochs      .
hist = model.fit(X_train, y_train, epochs=3000, verbose=2) 
```

```python
plt.plot(hist.history['loss'])
plt.plot(hist.history['accuracy'])
plt.xlabel('epoch')
plt.ylabel('loss')
plt.show()
```

 *w*  *b*  

```python
w, b = model.get_weights()
w =  w[0][0]
b = b[0]
print('w: ', w)
print('b: ', b)
```

```python
x_scale = mm_scaled.transform(X)
y_scale = mm_scaled.transform(y)
plt.scatter(x_scale, y_scale)
plt.plot(x_scale, w*x_scale+b, 'r')
plt.xlabel('scaled-height')
plt.ylabel('scaled-weight')
plt.show()
```

```python
input_height = 178.0

input_x = mm_scaled.transform(np.array([input_height]).reshape(-1, 1))
predict = model.predict(input_x)
predict = mm_scaled.inverse_transform(predict)

print(' {:.2f} kg '.format(predict[0][0]))
```
